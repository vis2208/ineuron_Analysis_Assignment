#  Drop rows with null

```python
df.dropna(inplace = True)
```

```python
df["RainToday"].isnull().sum()
```

```python
df["RainToday"].unique()  # categorical data check
```

```python
def yesno(rain):
    if rain == "Yes":
        return 1
    else:
        return 0
```

```python
df["RainToday"] = df["RainToday"].apply(yesno)   # converted Yes with 1 and No by 0
df["RainTomorrow"] = df["RainTomorrow"].apply(yesno)
```
