# IPL Matches 2008-2020

```python
df2 = pd.read_csv("IPL Matches 2008-2020.csv",parse_dates = ["date"])
df2[['city', 'player_of_match', 'venue',
       'team1', 'team2', 'toss_winner', 'toss_decision', 'winner','result',
    'eliminator', 'method', 'umpire1', 'umpire2']] = df2[['city', 'player_of_match', 'venue',
       'team1', 'team2', 'toss_winner', 'toss_decision', 'winner','result',
    'eliminator', 'method', 'umpire1', 'umpire2']].astype('category')
df2['result_margin'].fillna(0, inplace = True)
df2['result_margin'] = df2['result_margin'].astype('int')
df2['year'] = df2['date'].dt.year
df2.head(3)
```

# match details

```python

pd.options.display.max_rows = 1000
match_details = df2[['year','date','id','venue','city','team1','team2','toss_winner','toss_decision','winner','result','result_margin','player_of_match']]
match_details
```

# filled the NA city 

```python
def city(x):
    return x[0]
```

```python
df2["city"].fillna(df2[df2.city.isnull()]["venue"].str.split().apply(city), inplace = True)
```

# Merge both dataframe

```python
pd.options.display.max_columns = 100
df = df2.merge(df1,how = 'inner', on = ['id'])
df.head(3)
```

# Avearge Run by Player

```python
groupby_year_batsman = df.groupby(['year','batsman'])
```

```python
pd.options.display.max_rows = 104715
average_runby_player= groupby_year_batsman['total_runs'].mean().dropna().sort_values(ascending = False)
average_runby_player
```
